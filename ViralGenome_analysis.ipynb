```python
import sys
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import pandas as pd
import glob 
import math
import re
from collections import defaultdict
from matplotlib.font_manager import FontProperties
import subprocess
import scipy
from scipy.stats.stats import pearsonr 
%matplotlib inline
matplotlib.rcParams['savefig.dpi'] = 3 * matplotlib.rcParams['savefig.dpi']
```

```python
# Get raw data
global outfilepath
sampleName='jfh1v2'
outfilepath=os.getcwd() + '/results/%s/'%sampleName
print outfilepath
```

```python
def modifyName(filepath,newTag):
	# Useage: Modifies the filepath name. 
	# Input: File path of format <path>/<name>.fastq and a string to add to the name.
	# Output: Returns the modified path of type <old path>_<new modifier>.fastq
    head, tail = os.path.split(filepath)
    oldname = tail.split('.')[0]
    newName = head+"/"+oldname+"_"+newTag
    return newName

def runBowtie(fastqFiles,viralGenome):
    # Useage: Short read mapping to reference (hg19).
    # Input: Fastq files of replicates (not paired end).
    # Output: Path to samfile for each read.
    program = 'bowtie2'
    mappedReads=[]
    unMappedReads=[]
    print "Performing Bowtie..."
    # Parameters
    # In -k mode, Bowtie 2 searches for up to N distinct, valid alignments for each read.
    # N equals the integer specified with the -k parameter. 
    # That is, if -k 2 is specified, Bowtie 2 will search for at most 2 distinct alignments
    # See http://bowtie-bio.sourceforge.net/bowtie2/manual.shtml
    for infastq in fastqFiles:

        index = os.getcwd() + '/docs/%s/%s'%(viralGenome,viralGenome)
        outfile = modifyName(infastq,"mappedTo%s.sam"%viralGenome)
        unmapped = modifyName(infastq,"notMappedTo%s.fastq"%viralGenome)
        
        print "Input file:"
        print infastq 
        print 'Genome index:'
        print index
        print "Output file (mapped):"
        print outfile
        print "Output file (unmapped)"
        print unmapped
        # Remove -a
        proc = subprocess.Popen([program,'-x', index,'-U',infastq,'--un',unmapped,'-S',outfile],stdout=subprocess.PIPE,stderr=subprocess.PIPE)
        out, err = proc.communicate()
        result = out.decode()
        error = err.decode()
        print "Result : ",result 
        print "Error : ",error
        mappedReads = mappedReads + [outfile]
        unMappedReads = unMappedReads + [unmapped]
    return (mappedReads,unMappedReads)

def runSamtools(samfiles):
    # Useage: Samfile processing.
    # Input: Sam files from Bowtie mapping.
    # Output: Duplicate removed, sorted bedFiles.
    program = 'samtools'
    program2 = 'bamToBed'
    outBedFiles=[]
    
    for samfile in samfiles:
            
        # Convert to bamfile
        bamfile = samfile.replace('.sam', '.bam')  
        proc = subprocess.Popen( [program, 'view', '-bS', '-o', bamfile, samfile])
        proc.communicate()
        
        # Sort the bamfile and note that samtools sort adds the .bam handle
        bamfile_sort = bamfile.replace('.bam', '_sorted') 
        proc2 = subprocess.Popen( [program, 'sort', bamfile, bamfile_sort])
        proc2.communicate()
        
        # Convert to bedFile
        bedFile = bamfile_sort.replace('_sorted', '_withDupes.bed')
        outfh = open(bedFile, 'w')
        proc3 = subprocess.Popen( [program2,'-i', bamfile_sort+'.bam'],stdout=outfh)
        proc3.communicate()
        
        outBedFiles=outBedFiles+[bedFile]
        
    return outBedFiles

# Run bowtie
print "Mapping to viral index."  
trimmedReads5p=glob.glob(outfilepath+'*5ptrimmed.fastq')
mappedReads,unMappedReads=runBowtie(trimmedReads5p,'jfh1')

# Run Samtools
print "Process mapped data"  
mappedBedFiles=runSamtools(mappedReads)
```

```python
# Index
outfilepathjfh1=os.getcwd() + '/docs/jfh1/jfh1.fa'
jfh1_genome=np.genfromtxt(outfilepathjfh1,dtype='string')
jfh1_genome_size=len(jfh1_genome[1])

mappedBedFiles1=[]
sampleName='jfh1v2'
outfilepathjfh1=os.getcwd() + '/results/'+sampleName+'/'
mappedBedFiles2=glob.glob(outfilepathjfh1+'*mappedTojfh1_withDupes.bed')
sampleName='hcvwt'
outfilepathjfh1=os.getcwd() + '/results/'+sampleName+'/'
mappedBedFiles3=glob.glob(outfilepathjfh1+'*mappedTojfh1_withDupes.bed')
mappedBedFiles=mappedBedFiles1+mappedBedFiles2+mappedBedFiles3

# Global plotting stats
bins=np.array(range(0,jfh1_genome_size,5)) # Bin the genome by 5 bp
width=0.7*(bins[1]-bins[0]) # Bar width for plotting
center=(bins[:-1] + bins[1:])/2 # Center position of each bin

# Compute histogram for each
coverageHistograms=[]
for bedFile in mappedBedFiles:
    data=np.genfromtxt(bedFile,usecols=(1,),delimiter='\t',dtype='int') # Read the 5' coordinate (RT stop).
    hist,bins=np.histogram(data,bins=bins) # Compute histogram
    hist=hist/float(len(data)) # Normalize to total number of reads mapped to viral genome.
    coverageHistograms.append(hist)
```

```python
# Histograms to filter
HistToFilter=[0,1] # jfh1v2
Background=[2,3] # hcvwt
filteredHist=[]
# For each histogram
for i in HistToFilter:
    # Copy the values into temp object
    tempHist=coverageHistograms[i]
    # For each index in the histogram
    for index in range(0,len(tempHist)):
        # Check if background histograms (no infection) have counts
        if coverageHistograms[Background[0]][index] > 0 or coverageHistograms[Background[1]][index] > 0:
            # If so, then do not count
            tempHist[index]=0
    filteredHist.append(tempHist)
```

```python
# Scatter plot
geneType='HCV genome'
sampleName1='JFH-1 Rep1'
sampleName2='JFH-1 Rep2'
 
# Compute based upon histograms bins
bins=np.array(range(0,jfh1_genome_size,5)) # Bin the genome by 5 bp
width=0.7*(bins[1]-bins[0]) # Bar width for plotting
center=(bins[:-1] + bins[1:])/2 # Center position of each bin

# Annotation
UTR5end=340
UTR3start=9280

# Isolate UTR data from new JFH-1 histograms
boolarray=[element < UTR5end or element > UTR3start for element in center]
boolarray=np.array(boolarray,dtype=bool)
center=center[boolarray]
sample1=filteredHist[0][boolarray]
sample2=filteredHist[1][boolarray]

# Fit_fn is now a function which takes in x and returns an estimate for y
fit = np.polyfit(sample1,sample2,1)
fit_fn = np.poly1d(fit)
slope,intercept,r_value,p_value,std_err=scipy.stats.linregress(sample1,sample2)
print "r-squared:", r_value**2

fig1=plt.figure(1)

plt.subplot(1,2,2)
for i in range(0,len(center)):
    # Get position at the point
    position=center[i]
    # Annotate
    if position < UTR5end:
        color='blue'
        plt.scatter(sample1[i],sample2[i],color=color,alpha=0.35) 
    elif position > UTR3start:
        color='red'
        plt.scatter(sample1[i],sample2[i],color=color,alpha=0.35)
plt.plot(sample1,fit_fn(sample1),alpha=0.5,color='black')
plt.xlabel("Normalized gene read count: %s"%sampleName1,fontsize=5)
plt.ylabel("Normalized gene read count: %s"%sampleName2,fontsize=5)
plt.tick_params(axis='xticks',labelsize=5) 
plt.tick_params(axis='yticks',labelsize=5)
plt.xticks(rotation=90,fontsize=5)

#maxValx=0.012
#maxValy=0.008
#plt.xlim(-0.1*maxValx,maxValx)
#plt.ylim(-0.1*maxValy,maxValy)

plt.yticks(fontsize=5)
plt.tight_layout()	

fig1.savefig(outfilepath+'ViralScatter.png',format='png',bbox_inches='tight',dpi=200,pad_inches=0.5)
fig1.savefig(outfilepath+'ViralScatter.pdf',format='pdf',bbox_inches='tight',dpi=200,pad_inches=0.5)
```
